```python
import os

def rename_images_in_leaf_dirs(root_dir):
    for dirpath, dirnames, filenames in os.walk(root_dir):
        # Check if this is a leaf directory (no subdirs)
        if not dirnames:
            image_files = [f for f in filenames if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
            count = 1
            for img_file in image_files:
                old_path = os.path.join(dirpath, img_file)
                new_name = f"{count}{os.path.splitext(img_file)[1]}"
                new_path = os.path.join(dirpath, new_name)
                os.rename(old_path, new_path)
                print(f"Renamed: {old_path} -> {new_path}")
                count += 1
    print(count)
# Usage
rename_images_in_leaf_dirs("Dataset")
```
